```python
import numpy as np
scores = np.array([93,83,69,62,75,93,73,77,65,65,78,90,94,78,57,76,56,78,84,92])
n = len(scores)
sum_scores = scores.sum()
mean = scores.mean()
sumsq = ((scores-mean)**2).sum()
pop_var = sumsq / n
samp_var = sumsq / (n-1)
samp_std = np.sqrt(samp_var)
print(sum_scores, n, mean, sumsq, pop_var, samp_var, samp_std)
```

```python
import numpy as np
np.random.seed(42)
gen_scores = np.random.normal(loc=76.9, scale=11.968819138868602, size=20)
print(np.round(gen_scores,2))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(scores, bins=8, edgecolor='black')
plt.title("Real Exam Scores")
plt.subplot(1,2,2)
plt.hist(gen_scores, bins=8, edgecolor='black')
plt.title("Generated (Normal fit)")
plt.tight_layout()
plt.show()
```
